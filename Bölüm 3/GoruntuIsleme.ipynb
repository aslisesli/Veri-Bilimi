# YZ TANIMA

```python
yuz=cv2.imread("yuz.jpeg")
plt.imshow(yuz)
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
eye_cascade=cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
faces=face_cascade.detectMultiScale(yuz,scaleFactor=1.075,minNeighbors=5,minSize=(15,15)) # resim zerinde yzleri cerceveliyoruz ve kordinatlarn faces degiskeniinde tutuyoruz
```

```python
for (x,y,w,h) in faces:                               # resimde birden fazla yuz oldugu iin for dngs kuruyoruz yuzler iin diktorgen olustuuryoruz
    cv2.rectangle(yuz,(x,y),(x+w,y+h),(255,0,0),2)  
    roi=yuz[y:y+h,x:x+w]
    eyes=eye_cascade.detectMultiScale(roi)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi,(ex,ey),(ex+ew,ey+eh),(0,255,0),1)
```

```python
print("yuz bulundu! bulunan yuz says:",format(len(faces)))
```

```python
plt.figure(figsize=(8,12))
plt.imshow(yuz)
plt.show()
```

# perspektif

```python
img = cv2.imread('suduko.jpg')
pts1 = np.float32([[25,25],[185,25],[10,195],[195,195]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
M = cv2.getPerspectiveTransform(pts1,pts2)
pers = cv2.warpPerspective(img,M,(200,200))
plt.subplot(121), plt.imshow(img),plt.title('Orijinal')
plt.subplot(122), plt.imshow(pers),plt.title('Perspektif')
plt.show()
```
